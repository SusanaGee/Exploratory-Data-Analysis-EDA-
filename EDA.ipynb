## 🔍 **¿Pero... qué es un EDA?**

El término **EDA** proviene de sus siglas en inglés: _Exploratory Data Analysis_ 📊.  

Como su nombre lo indica, se trata de explorar y estudiar a profundidad nuestros datos, realizar ajustes preliminares y extraer una primera serie de conclusiones 🧪💡. Todo esto **antes** de aplicar cualquier método avanzado, como técnicas de *machine learning* 🤖.

El EDA es, en esencia, el primer paso para comprender qué está ocurriendo dentro de nuestro conjunto de datos y descubrir patrones, tendencias y posibles problemas. 🛠️📈

Este dataset contiene un total de **12 columnas**, que describimos a continuación:

#### 🧩 **Paso #2: Identificar datos faltantes**

Cuando estamos realizando un **Análisis Exploratorio de Datos (EDA)**, es crucial identificar si tenemos **datos faltantes** ⚠️.  
Estos pueden afectar sustancialmente los resultados que obtengamos en procesos más avanzados, como modelos de *machine learning* o análisis estadísticos.

Detectar valores nulos desde el inicio nos permite tomar decisiones informadas, como:

- ¿Deberíamos **imputar** (rellenar) los valores faltantes? 🧮
- ¿O es mejor **eliminarlos** del análisis? 🗑️

```python
# Para encontrar los datos faltantes en nuestro dataset hay mucho métodos,
# sin embargo, aquí solo te explicaré uno pues el que mejor me funciona ya que 
# es mucho más visual

df.isna().sum().reset_index().rename(columns={"index":"Variable", 0:"Valores faltantes"})
```

Como podemos notar, tenemos 3 variables con valores faltantes. Antes de proceder con el resto del análisis, revisaremos cada una de ellas y estudiaremos a qué alternativa podemos recurrir para cada una.


##### **Age** 👶📊

Si bien es una variable que tiene una cantidad importante de datos faltantes, estos los podemos imputar con la **media**, de tal manera que no perdamos la información disponible en las otras variables de estas observaciones.


```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df.isna().sum()
```

##### **Cabin** 🛏️🚪

Por su parte, la variable `Cabin` tiene un número más significativo de valores faltantes, por lo que no se considera viable aplicar ningún método de imputación y, en su lugar, se opta por eliminar la variable.


```python
df.drop(columns='Cabin', inplace=True)
df.isna().sum()
```

##### **Embarked** ⚓👀

Para concluir esta sección, está el caso de la variable `Embarked`, la cual tiene solo dos valores faltantes, por lo que se considera prudente mirar detenidamente ambas observaciones y poder determinar el tratamiento óptimo.

```python
df[df['Embarked'].isna()]
```

Tal y como se observa, ambos pasajeros tienen el mismo número de ticket, lo cual no resulta coherente, por lo que se opta por eliminar ambas observaciones.


```python
df.dropna(inplace=True)
df.isna().sum()
```

De manera similar debemos corroborar si hay más duplicados, diferente al ya eliminado, de lo contrario podremos avanzar con nuestro análisis.

```python
df.duplicated().sum()
```

Como indican los resultados, en este caso nuestro dataset no cuenta con más valores duplicados.

#### **Tabla de Frecuencia** 📊

Como su nombre lo indica, este tipo de tablas nos permiten identificar, de manera porcentual, con qué frecuencia un pasajero —en este caso— pertenecía a alguna de las tres clases que había en el barco. 🎟️⚓


```python
class_ft=df["Pclass"].value_counts().reset_index()
class_ft["percent"]=100*class_ft["count"]/class_ft["count"].sum()
class_ft["cum_percent"]=class_ft["percent"].cumsum()
class_ft
```

A partir de los resultados, podemos concluir que: un **55.23%** de los pasajeros pertenecían a la **tercera clase** (albergando el mayor porcentaje de los pasajeros) 🚶‍♂️🚶‍♀️, un **24.07%** a la **primera clase** (siendo la segunda clase con mayor cantidad de pasajeros) 🎩🍷, y finalmente un **20.69%** pertenecían a la **clase media** 👜🧥.


Los resultados presentados arriba podemos visualizarlos en un `Barplot` con el siguiente código:

```python
class_ft.plot(
    kind="barh",
    figsize=(6,4),
    grid=True,
    x="Pclass", y="percent",
    title="Pasajeros por clase",
    xlabel="Porcentaje",
    color="pink"
)
```

#### 🔄 **Análisis de Correlaciones entre Variables Numéricas** 📊🔍

Una vez hemos comprendido de manera aislada el comportamiento de algunas de las variables más relevantes, ahora nos enfocaremos en entender **cómo interactúan entre sí las variables numéricas**. 🤝🔢

Nuestro objetivo será identificar si existe algún nivel de **correlación** entre ellas y, en caso de encontrar evidencia de dicha relación, **analizar qué implicaciones puede tener** en nuestro análisis. 🧠📈

Para cumplir con este propósito, utilizaremos el siguiente fragmento de código: ⬇️💻


```python
variables_relevantes = df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch','Fare' ]]
paleta_colores = ["#ECCEF5", "#E2A9F3", "#CC2EFA", "#6A0888", "#3B0B39"]
correlacion = variables_relevantes.corr()


sns.heatmap(correlacion, annot=True, cmap= paleta_colores, vmin=-1, vmax=1, square=True, fmt=".2f")
plt.title('Mapa de Calor o Correlación')
plt.show()
```


Para poder analizar adecuadamente este gráfico, es crucial comprender primero **tres aspectos básicos** sobre este tipo de representaciones:

1. ➕ **Correlación positiva fuerte**: Cuando los valores están cerca de **1**, significa que existe una fuerte relación positiva entre las variables. Es decir, **si una variable tiende a aumentar, la otra también lo hará**, en una proporción similar.

2. ➖ **Correlación negativa fuerte**: Si los valores están cerca de **-1**, indica una fuerte relación negativa. Esto quiere decir que **cuando una variable aumenta, la otra tiende a disminuir**, también en proporciones similares.

3. ➗ **Sin correlación**: Cuando los valores están cerca de **0**, no hay evidencia de una relación lineal entre las variables (ya sea positiva o negativa).


##### 🧠 Análisis Específico de Algunas Variables Clave:

✅ `Survived`
Se observa una **correlación negativa moderada con `Pclass`**, lo cual sugiere que los pasajeros de **clases inferiores (2 y 3)** tendieron a sobrevivir menos. Esto concuerda con análisis previos, ya que fueron principalmente **mujeres y niños de primera clase** quienes lograron sobrevivir en mayor proporción.  

Además, se evidencia una **correlación positiva moderada con `Fare`**, indicando que las personas que pagaron más por su tiquete tuvieron mayores probabilidades de sobrevivir, respaldando así lo observado respecto a la variable `Pclass`.

🛏️ `Pclass`
Al analizar esta variable y sus correlaciones, encontramos información que **refuerza lo observado con `Survived`**. Específicamente, se identifica una **correlación negativa fuerte con `Fare`**, confirmando que los pasajeros de **primera clase pagaron más** por su tiquete.  

Sumado a esto, se encuentra que los pasajeros de primera clase también eran, en promedio, de **mayor edad**, lo cual complementa el análisis previo.

 👨‍👩‍👧 `SibSp` y `Parch`

#### 🛟 **Análisis de Supervivencia Condensado** 📊

Para concluir esta aventura, descifrando los datos más importantes del naufragio más famoso de la historia, aún podría quedarnos una duda.  

Si bien durante el análisis de correlaciones identificamos algunos **patrones en la supervivencia de los pasajeros**, así como **variables que pudieron haber influido fuertemente en sus probabilidades de sobrevivir**, todavía **desconocemos con claridad la proporción exacta entre quienes sobrevivieron y quienes no**.  

Para visualizar este hallazgo final, te propongo el siguiente **gráfico de torta 🥧**, que nos permitirá obtener una **vista general y directa** de este importante aspecto de nuestro análisis.


```python

conteo = df['Survived'].value_counts()
etiquetas = ['No sobrevivió', 'Sobrevivió']
colores = ['#C8A2C8', '#800020']

plt.figure(figsize=(6, 6))
plt.pie(conteo, labels=etiquetas, autopct='%1.1f%%', startangle=90, colors=colores, shadow=True)
plt.title('Distribución de Supervivencia', fontsize=14)
plt.axis('equal')

plt.show()
```

Por medio de este gráfico se identifica que, efectivamente (tal y como plantean los historiadores), menos de la mitad de los pasajeros lograron sobrevivir al naufragio más famoso de la historia. Solo el 38.2% de los tripulantes lograron subir al buque de rescate, el *Carpathia*, mientras que un 61.8% perecieron en las heladas aguas del Atlántico Norte ❄️🚢.
