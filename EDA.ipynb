#### üß© **Paso #2: Identificar datos faltantes**

Cuando estamos realizando un **An√°lisis Exploratorio de Datos (EDA)**, es crucial identificar si tenemos **datos faltantes** ‚ö†Ô∏è.  
Estos pueden afectar sustancialmente los resultados que obtengamos en procesos m√°s avanzados, como modelos de *machine learning* o an√°lisis estad√≠sticos.

Detectar valores nulos desde el inicio nos permite tomar decisiones informadas, como:

- ¬øDeber√≠amos **imputar** (rellenar) los valores faltantes? üßÆ
- ¬øO es mejor **eliminarlos** del an√°lisis? üóëÔ∏è

```python
# Para encontrar los datos faltantes en nuestro dataset hay mucho m√©todos,
# sin embargo, aqu√≠ solo te explicar√© uno pues el que mejor me funciona ya que 
# es mucho m√°s visual

df.isna().sum().reset_index().rename(columns={"index":"Variable", 0:"Valores faltantes"})
```

Como podemos notar, tenemos 3 variables con valores faltantes. Antes de proceder con el resto del an√°lisis, revisaremos cada una de ellas y estudiaremos a qu√© alternativa podemos recurrir para cada una.


##### **Age** üë∂üìä

Si bien es una variable que tiene una cantidad importante de datos faltantes, estos los podemos imputar con la **media**, de tal manera que no perdamos la informaci√≥n disponible en las otras variables de estas observaciones.


```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df.isna().sum()
```

##### **Cabin** üõèÔ∏èüö™

Por su parte, la variable `Cabin` tiene un n√∫mero m√°s significativo de valores faltantes, por lo que no se considera viable aplicar ning√∫n m√©todo de imputaci√≥n y, en su lugar, se opta por eliminar la variable.


```python
df.drop(columns='Cabin', inplace=True)
df.isna().sum()
```

##### **Embarked** ‚öìüëÄ

Para concluir esta secci√≥n, est√° el caso de la variable `Embarked`, la cual tiene solo dos valores faltantes, por lo que se considera prudente mirar detenidamente ambas observaciones y poder determinar el tratamiento √≥ptimo.

```python
df[df['Embarked'].isna()]
```

#### **Tabla de Frecuencia** üìä

Como su nombre lo indica, este tipo de tablas nos permiten identificar, de manera porcentual, con qu√© frecuencia un pasajero ‚Äîen este caso‚Äî pertenec√≠a a alguna de las tres clases que hab√≠a en el barco. üéüÔ∏è‚öì


```python
class_ft=df["Pclass"].value_counts().reset_index()
class_ft["percent"]=100*class_ft["count"]/class_ft["count"].sum()
class_ft["cum_percent"]=class_ft["percent"].cumsum()
class_ft
```

A partir de los resultados, podemos concluir que: un **55.23%** de los pasajeros pertenec√≠an a la **tercera clase** (albergando el mayor porcentaje de los pasajeros) üö∂‚Äç‚ôÇÔ∏èüö∂‚Äç‚ôÄÔ∏è, un **24.07%** a la **primera clase** (siendo la segunda clase con mayor cantidad de pasajeros) üé©üç∑, y finalmente un **20.69%** pertenec√≠an a la **clase media** üëúüß•.


Los resultados presentados arriba podemos visualizarlos en un `Barplot` con el siguiente c√≥digo:

```python
class_ft.plot(
    kind="barh",
    figsize=(6,4),
    grid=True,
    x="Pclass", y="percent",
    title="Pasajeros por clase",
    xlabel="Porcentaje",
    color="pink"
)
```

#### üîÑ **An√°lisis de Correlaciones entre Variables Num√©ricas** üìäüîç

Una vez hemos comprendido de manera aislada el comportamiento de algunas de las variables m√°s relevantes, ahora nos enfocaremos en entender **c√≥mo interact√∫an entre s√≠ las variables num√©ricas**. ü§ùüî¢

Nuestro objetivo ser√° identificar si existe alg√∫n nivel de **correlaci√≥n** entre ellas y, en caso de encontrar evidencia de dicha relaci√≥n, **analizar qu√© implicaciones puede tener** en nuestro an√°lisis. üß†üìà

Para cumplir con este prop√≥sito, utilizaremos el siguiente fragmento de c√≥digo: ‚¨áÔ∏èüíª


```python
variables_relevantes = df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch','Fare' ]]
paleta_colores = ["#ECCEF5", "#E2A9F3", "#CC2EFA", "#6A0888", "#3B0B39"]
correlacion = variables_relevantes.corr()


sns.heatmap(correlacion, annot=True, cmap= paleta_colores, vmin=-1, vmax=1, square=True, fmt=".2f")
plt.title('Mapa de Calor o Correlaci√≥n')
plt.show()
```


Para poder analizar adecuadamente este gr√°fico, es crucial comprender primero **tres aspectos b√°sicos** sobre este tipo de representaciones:

1. ‚ûï **Correlaci√≥n positiva fuerte**: Cuando los valores est√°n cerca de **1**, significa que existe una fuerte relaci√≥n positiva entre las variables. Es decir, **si una variable tiende a aumentar, la otra tambi√©n lo har√°**, en una proporci√≥n similar.

2. ‚ûñ **Correlaci√≥n negativa fuerte**: Si los valores est√°n cerca de **-1**, indica una fuerte relaci√≥n negativa. Esto quiere decir que **cuando una variable aumenta, la otra tiende a disminuir**, tambi√©n en proporciones similares.

3. ‚ûó **Sin correlaci√≥n**: Cuando los valores est√°n cerca de **0**, no hay evidencia de una relaci√≥n lineal entre las variables (ya sea positiva o negativa).
